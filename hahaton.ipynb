```python
def find_common_columns(file_list):
    # Создаем пустой список для хранения столбцов каждой таблицы
    columns_list = []
    
    # Проходим по каждому пути к таблице
    for file_path in file_list:
        # Загружаем таблицу
        table = pd.read_csv(file_path)
        
        # Получаем множество столбцов таблицы
        columns = set(table.columns)
        
        # Добавляем множество столбцов в список
        columns_list.append(columns)
    
    # Используем метод intersection для нахождения общих столбцов
    common_columns = set.intersection(*columns_list)
    
    # Выводим общие столбцы
    print("Общие столбцы для всех таблиц:")
    print(common_columns)

# Пример вызова функции
find_common_columns(file_list)
```

```python
def count_columns_occurrence(file_list, column1, column2):
    # Инициализируем счетчики для каждого столбца
    count_column1 = 0
    count_column2 = 0
    
    # Проходим по каждому пути к таблице
    for file_path in file_list:
        # Загружаем таблицу
        table = pd.read_csv(file_path)
        
        # Проверяем наличие столбца 1
        if column1 in table.columns:
            count_column1 += 1
        
        # Проверяем наличие столбца 2
        if column2 in table.columns:
            count_column2 += 1
    
    # Выводим результаты
    print(f"Столбец {column1} содержится в {count_column1} таблицах.")
    print(f"Столбец {column2} содержится в {count_column2} таблицах.")

# Пример вызова функции
column1 = 'ФИО участников'
column2 = 'ФИО'
count_columns_occurrence(file_list, column1, column2)
```

```python
import os
import pandas as pd

col = ['ФИО', 'Название команды', 'Количество участников', 'Компетенция', 'Баллы, %']

df = pd.DataFrame(columns=col)

def split_names(text):
    names = text.split(";")
    names = [name.strip() for name in names]
    return names


for filename in os.listdir(directory):
    a = pd.read_csv(directory+filename)
    
    if 'ФИО' in a.columns.to_list():
        
        for i in range(a.index.to_list()[-1]+1):
            row = a.iloc[i]
            fio = row['ФИО']
            komp = row['Компетенция']
            points = row['Баллы, %']
            
            new_row = {'ФИО': fio, 'Название команды': fio, 'Количество участников': 1,'Компетенция': komp,'Баллы, %': points}
            df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

    elif 'ФИО участников' in a.columns.to_list():
        
        for i in range(a.index.to_list()[-1]+1):
            
            row = a.iloc[i]
            team = split_names(row['ФИО участников'])
            memb_count = len(team)
            team_name = row['Название команды']
            komp = row['Компетенция']
            points = row['Баллы, %']
            
            for name in team:
                new_row = {'ФИО': name, 'Название команды': team_name, 'Количество участников': memb_count,'Компетенция': komp,'Баллы, %': points}
                df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

    else:
        print(filename, "Ошибка")
```

```python
def calculate_age(birth_date):
    if pd.isnull(birth_date):  # Если значение ячейки пустое, оставляем его без изменений
        return birth_date
    
    current_date = datetime.now().date()
    
    # Поддерживаемые форматы даты
    date_formats = [
        "%d.%m.%Y",
        "%d.%m.%y",
        "%Y-%m-%d",
        "-%m-%d",
        "%m-%d-%Y"
    ]
    
    for date_format in date_formats:
        try:
            birth_date = datetime.strptime(birth_date, date_format).date()
            break  # Если формат успешно найден, выходим из цикла
        except ValueError:
            pass
    
    # Если формат даты не найден, возвращаем исходное значение
    if isinstance(birth_date, str):
        return birth_date
    
    age = current_date.year - birth_date.year
    # Проверяем, был ли уже день рождения в текущем году
    if (birth_date.month, birth_date.day) > (current_date.month, current_date.day):
        age -= 1
    
    return age

merged['Дата рождения'] = merged['Дата рождения'].apply(calculate_age)
```

```python
# Словарь для сопоставления названия месяца с числовым значением
month_mapping = {
    'январь': '01',
    'февраль': '02',
    'март': '03',
    'апрель': '04',
    'май': '05',
    'июнь': '06',
    'июль': '07',
    'август': '08',
    'сентябрь': '09',
    'октябрь': '10',
    'ноябрь': '11',
    'декабрь': '12',
    'янв': '01',
    'фев': '02',
    'мар': '03',
    'апр': '04',
    'май': '05',
    'июн': '06',
    'июл': '07',
    'авг': '08',
    'сен': '09',
    'сент': '09',
    'окт': '10',
    'ноя': '11',
    'дек': '12'
}

# Определение функции для вычисления количества месяцев
def months_until_current(date_string):
    current_date = datetime.now().date()
    if date_string == '':
        return date_string

    if '.' in date_string:
        parts = date_string.split('.')
        if len(parts) == 3:
            try:
                day = int(parts[0])
                month = parts[1].lower()
                year = int(parts[2])
                if 1 <= day <= 31 and month in month_mapping and 1900 <= year <= current_date.year:
                    month_number = month_mapping[month]
                    date_string = f"{day:02d}.{month_number}.{year:02d}"
                    date_format = "%d.%m.%y"
                else:
                    return date_string
            except ValueError:
                return date_string
        elif len(parts) == 2:
            month = parts[0].lower()
            year = parts[1]
            if month in month_mapping and year.isdigit() and 1900 <= int(year) <= current_date.year:
                month_number = month_mapping[month]
                date_string = f"{month_number:02d}.01.{int(year):02d}"
                date_format = "%m.%d.%y"
            else:
                return date_string
        elif len(parts) == 1:
            year = parts[0]
            if year.isdigit() and 1900 <= int(year) <= current_date.year:
                date_string = f"01.01.{int(year):02d}"
                date_format = "%m.%d.%y"
            else:
                return date_string
        else:
            return date_string
    elif '-' in date_string:
        parts = date_string.split('-')
        if len(parts) == 3:
            try:
                date_format = "%Y-%m-%d"
            except ValueError:
                return date_string
        elif len(parts) == 2:
            if parts[0].isdigit() and parts[1].isdigit():
                month = int(parts[0])
                year = int(parts[1])
                if 1 <= month <= 12 and 1900 <= year <= current_date.year:
                    date_string = f"{month:02d}.01.{year:02d}"
                    date_format = "%m.%d.%y"
                else:
                    return date_string
            else:
                return date_string
        else:
            return date_string
    elif date_string.isdigit():
        year = int(date_string)
        if 1900 <= year <= current_date.year:
            date_string = f"01.01.{year:02d}"
            date_format = "%m.%d.%y"
        else:
            return date_string
    else:
        parts = date_string.split()
        if len(parts) == 2:
            month = parts[0].lower()
            year = parts[1]
            if month not in month_mapping:
                return date_string
            month_number = month_mapping[month]
            date_string = f"{month_number:02d}.01.{year}"
            date_format = "%m.%d.%Y"
        else:
            return date_string

    try:
        date_object = datetime.strptime(date_string, date_format).date()
    except ValueError:
        return date_string

    delta = relativedelta(current_date, date_object)
    months = delta.years * 12 + delta.months
    return months

# Фильтрация строк по формату даты
valid_dates = merged['Начало трудового стажа'].apply(lambda x: isinstance(x, str))
merged = merged[valid_dates]


merged['Начало трудового стажа'] = merged['Начало трудового стажа'].apply(months_until_current)
```
